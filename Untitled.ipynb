```python
from sklearn.utils import resample
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
```

```python
y_col = ['x24h_cond_loc','x48h_cond_loc']
model_col = []

acc_col = []
recall_col = []
precision_col = []
f1_col = []

plant_col = []
hr_col = []
loc_col = []
model_dict = {}
for y in y_col:
    for plant in (1,2):
        for loc in (1,2,3):
            df_to_split = df_plant_train[(df_plant_train['plant'] == plant)&(df_plant_train['loc'] == loc)].drop(['mea_ddhr','plant','loc'],axis=1)

            plant_x_train = df_to_split.loc[:,~df_to_split.columns.isin(y_col)]
            plant_y_train = df_to_split.loc[:,[y]]
            
            x_train,x_test, y_train,y_test = train_test_split(plant_x_train, plant_y_train, test_size = 0.33, random_state = 777)
            
            
            df = pd.concat([x_train,y_train],axis=1)
            
            df_major = df[df[y]==0]
            df_minor = df[df[y]==1]

            df_minor_up = resample(df_minor,
                                   replace=True,
                                   n_samples = len(df_major),
                                   random_state = 777)

            df_upsampled = pd.concat([df_major,df_minor_up])

            x_train = df_upsampled.loc[:,~df_upsampled.columns.isin(y_col)]
            y_train = df_upsampled.loc[:,[y]]

            



            df = pd.merge(x_train,y_train,left_index=True, right_index=True)


            lr_model = LogisticRegression()
            lr_model.fit(x_train,y_train[y])
            
            y_pred = lr_model.predict(x_test)
            y_true = y_test[y]

            acc = accuracy_score(y_pred,y_true)
            recall = recall_score(y_pred,y_true)
            precision = precision_score(y_pred,y_true)
            f1 = f1_score(y_pred,y_true)

            acc_col.append(acc)
            recall_col.append(recall)
            precision_col.append(precision)
            f1_col.append(f1)


            if plant == 1:
                plant_col.append('1')
            else:
                plant_col.append('2')

            if 'x24' in y:
                hr_col.append('x24')
            else:
                hr_col.append('x48')

            if loc  == 1:
                loc_col.append('1')

            elif loc == 2:
                loc_col.append('2')
            else:
                loc_col.append('3')

            model_col.append(lr_model)
            

df_result_lr = pd.DataFrame([plant_col,hr_col,loc_col,acc_col,recall_col,precision_col,f1_col],index=['plant','hr','loc','acc','recall','precision','f1']).T      
print('done')
```

```python
df_result_lr.acc = df_result_lr.acc.astype('float')
df_result_lr.recall = df_result_lr.recall.astype('float')
df_result_lr.precision = df_result_lr.precision.astype('float')
df_result_lr.f1 = df_result_lr.f1.astype('float')
```
